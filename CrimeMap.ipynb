```python
import pandas as pd
df = pd.read_csv('crime_sample.csv')
df.head()
```

```python
df = df.dropna(subset=['Latitude', 'Longitude'])
df = df[(df['Latitude'].between(-90, 90)) & (df['Longitude'].between(-180, 180))]
df.shape
```

```python
import folium

map_center = [37.7749, -122.4194]
crime_map = folium.Map(location=map_center, zoom_start=13)
crime_map
```

```python
for idx, row in df.iterrows():
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=3,
        color='red',
        fill=True,
        fill_opacity=0.5
    ).add_to(crime_map)

crime_map
```

```python
from folium.plugins import HeatMap

heat_map = folium.Map(location=map_center, zoom_start=13)
HeatMap(data=df[['Latitude', 'Longitude']]).add_to(heat_map)
heat_map
```
